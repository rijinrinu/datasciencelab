```python
import numpy as np

# 1. Add arr1 and arr2
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([6, 7, 8, 9, 10])
result_add = np.add(arr1, arr2)

# 2. Multiply arr1 and arr2
result_multiply = np.multiply(arr1, arr2)

# 3. Calculate the mean of result_add
mean_result_add = np.mean(result_add)

# 4. Find the maximum value in result_multiply
max_result_multiply = np.max(result_multiply)

print("result_add: ", result_add)
print("result_multiply: ", result_multiply)
print("Mean of result_add: ", mean_result_add)
print("Maximum of result_multiply: ", max_result_multiply)
```

```python
import numpy as np

import matplotlib.pyplot as plt

x=np.arange(0,3*np.pi,0.1)

y_sin=np.sin(x)

y_cos=np.cos(x)

plt.plot(x,y_sin)

plt.plot(x,y_cos)

plt.xlabel('x axis label')

plt.ylabel('y axis label')

plt.title('since and cosine')

plt.legend(['since','cosine'])
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
iris=load_iris()
x=iris.data
y=iris.target
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
c_knn=KNeighborsClassifier(n_neighbors=3)
c_knn.fit(x_train,y_train)
y_pred=c_knn.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
sample=[[2,2,2,2]]
pred=c_knn.predict(sample)
pred_v=[iris.target_names[p] for p in pred]
print(pred_v)
```

```python
from sklearn.datasets import load_iris
from sklearn import metrics
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

iris=load_iris()
x=iris.data
y=iris.target
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
clf=DecisionTreeClassifier()
clf=clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
print("accuracy:",metrics.accuracy_score(y_test,y_pred))
plt.figure(figsize=(15,15))
tree.plot_tree(clf,fontsize=10,filled=True,rounded=True,class_names=iris.target_names,feature_names=iris.feature_names)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets,linear_model
from sklearn.metrics import mean_squared_error,r2_score

diabetes_X,diabetes_y=datasets.load_diabetes(return_X_y=True)
diabetes_X=diabetes_X[:,np.newaxis,2]
diabetes_X.shape
diabetes_y.shape
diabetes_X_train=diabetes_X[:-20]
diabetes_X_test=diabetes_X[-20:]
diabetes_y_train=diabetes_y[:-20]
diabetes_y_test=diabetes_y[-20:]

regr=linear_model.LinearRegression()

regr.fit(diabetes_X_train,diabetes_y_train)

diabetes_y_pred=regr.predict(diabetes_X_test)

print('coefficients:\n',regr.coef_)
print('mean squared error:%2f'%mean_squared_error(diabetes_y_test,diabetes_y_pred))
print('coefficient determination:%2f'%r2_score(diabetes_y_test,diabetes_y_pred))

plt.scatter(diabetes_X_test,diabetes_y_test,color='black')
plt.plot(diabetes_X_test,diabetes_y_pred,color='blue',linewidth='3')
plt.xticks(())
plt.yticks(())
plt.show()
```
